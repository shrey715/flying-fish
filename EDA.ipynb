# Exploratory Data Analysis - Auto Insurance Churn Dataset

This notebook performs comprehensive EDA on the auto insurance churn dataset, analyzing customer demographics, address information, and termination patterns.

## 2. Load the Datasets

```python
# Change to the correct working directory
import os
notebook_path = '/home/abhishekg/Documents/flying-fish'
os.chdir(notebook_path)
print(f"Changed working directory to: {os.getcwd()}")
print(f"Files in archive: {os.listdir('archive')}")
```

```python
# Define the data directory
data_dir = Path('archive')

# Load all datasets
address_df = pd.read_csv(data_dir / 'address.csv')
customer_df = pd.read_csv(data_dir / 'customer.csv')
demographic_df = pd.read_csv(data_dir / 'demographic.csv')
termination_df = pd.read_csv(data_dir / 'termination.csv')
autoinsurance_df = pd.read_csv(data_dir / 'autoinsurance_churn.csv')
```

## 3. Initial Data Exploration

```python
# Display first few rows of each dataset
print("=" * 80)
print("ADDRESS DATASET")
print("=" * 80)
display(address_df.head())
print("\n")

print("=" * 80)
print("CUSTOMER DATASET")
print("=" * 80)
display(customer_df.head())
print("\n")

print("=" * 80)
print("DEMOGRAPHIC DATASET")
print("=" * 80)
display(demographic_df.head())
print("\n")

print("=" * 80)
print("TERMINATION DATASET")
print("=" * 80)
display(termination_df.head())
print("\n")

print("=" * 80)
print("AUTO INSURANCE CHURN DATASET")
print("=" * 80)
display(autoinsurance_df.head())
```

### 3.1 Data Structure and Types

```python
# Check data types and info for each dataset
datasets = {
    'Address': address_df,
    'Customer': customer_df,
    'Demographic': demographic_df,
    'Termination': termination_df,
    'Auto Insurance': autoinsurance_df
}

for name, df in datasets.items():
    print("=" * 80)
    print(f"{name.upper()} DATASET INFO")
    print("=" * 80)
    print(df.info())
    print("\n")
```

### 3.2 Missing Values Analysis

```python
# Check for missing values in each dataset
for name, df in datasets.items():
    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    missing_df = pd.DataFrame({
        'Missing Count': missing,
        'Percentage': missing_pct
    }).sort_values('Missing Count', ascending=False)
    
    if missing.sum() > 0:
        print(f"\n{name.upper()} - Missing Values:")
        print(missing_df[missing_df['Missing Count'] > 0])
    else:
        print(f"\n{name.upper()}: No missing values!")
    print("-" * 50)
```

### 3.3 Statistical Summary

```python
# Statistical summary for numerical columns
for name, df in datasets.items():
    print("=" * 80)
    print(f"{name.upper()} - Statistical Summary")
    print("=" * 80)
    display(df.describe())
    print("\n")
```

## 4. Merge Datasets for Comprehensive Analysis

```python
# Merge all datasets to create a comprehensive view
# First, let's check the common columns for merging
print("Columns in each dataset:")
for name, df in datasets.items():
    print(f"\n{name}: {list(df.columns)}")

# Perform merges (assuming there's a customer ID or similar key)
# We'll merge based on common columns
merged_df = customer_df.copy()

# Merge with demographic if they share a common key
if 'customer_id' in customer_df.columns or any('id' in col.lower() for col in customer_df.columns):
    common_cols = set(customer_df.columns).intersection(set(demographic_df.columns))
    if common_cols:
        merge_key = list(common_cols)[0]
        merged_df = merged_df.merge(demographic_df, on=merge_key, how='left', suffixes=('', '_demo'))
        print(f"Merged with demographic on: {merge_key}")

print(f"\nMerged dataset shape: {merged_df.shape}")
print(f"Columns: {list(merged_df.columns)}")
```

## 5. Auto Insurance Churn Analysis

```python
# Detailed exploration of auto insurance churn dataset
print("AUTO INSURANCE CHURN DATASET - Detailed Analysis")
print("=" * 80)
print(f"Dataset shape: {autoinsurance_df.shape}")
print(f"\nColumn names and types:")
print(autoinsurance_df.dtypes)
print(f"\nFirst 10 rows:")
display(autoinsurance_df.head(10))

# Check for categorical columns
categorical_cols = autoinsurance_df.select_dtypes(include=['object']).columns
numerical_cols = autoinsurance_df.select_dtypes(include=['int64', 'float64']).columns

print(f"\nCategorical columns: {list(categorical_cols)}")
print(f"Numerical columns: {list(numerical_cols)}")
```

### 5.1 Categorical Variables Distribution

```python
# Analyze categorical variables
for col in categorical_cols:
    print(f"\n{col.upper()} - Value Counts:")
    print("-" * 50)
    value_counts = autoinsurance_df[col].value_counts()
    print(value_counts)
    print(f"\nUnique values: {autoinsurance_df[col].nunique()}")
```

### 5.2 Visualizations - Distribution of Key Variables

```python
# Visualize categorical variables
if len(categorical_cols) > 0:
    n_cols = min(3, len(categorical_cols))
    n_rows = (len(categorical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if len(categorical_cols) > 1 else [axes]
    
    for idx, col in enumerate(categorical_cols):
        if idx < len(axes):
            value_counts = autoinsurance_df[col].value_counts()
            axes[idx].bar(range(len(value_counts)), value_counts.values)
            axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('Count')
            axes[idx].set_xticks(range(len(value_counts)))
            axes[idx].set_xticklabels(value_counts.index, rotation=45, ha='right')
            axes[idx].grid(axis='y', alpha=0.3)
    
    # Hide empty subplots
    for idx in range(len(categorical_cols), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("No categorical columns to visualize")
```

### 5.3 Numerical Variables Distribution

```python
# Visualize numerical variables with histograms
if len(numerical_cols) > 0:
    n_cols = min(3, len(numerical_cols))
    n_rows = (len(numerical_cols) + n_cols - 1) // n_cols
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5*n_rows))
    axes = axes.flatten() if len(numerical_cols) > 1 else [axes]
    
    for idx, col in enumerate(numerical_cols):
        if idx < len(axes):
            axes[idx].hist(autoinsurance_df[col].dropna(), bins=30, edgecolor='black', alpha=0.7)
            axes[idx].set_title(f'Distribution of {col}', fontsize=12, fontweight='bold')
            axes[idx].set_xlabel(col)
            axes[idx].set_ylabel('Frequency')
            axes[idx].grid(axis='y', alpha=0.3)
    
    # Hide empty subplots
    for idx in range(len(numerical_cols), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()
else:
    print("No numerical columns to visualize")
```

### 5.4 Correlation Analysis

```python
# Correlation matrix for numerical variables
if len(numerical_cols) > 1:
    plt.figure(figsize=(12, 10))
    correlation_matrix = autoinsurance_df[numerical_cols].corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
                fmt='.2f', square=True, linewidths=1)
    plt.title('Correlation Matrix - Auto Insurance Dataset', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Find highly correlated features
    print("\nHighly Correlated Features (|correlation| > 0.7):")
    print("=" * 60)
    for i in range(len(correlation_matrix.columns)):
        for j in range(i+1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) > 0.7:
                print(f"{correlation_matrix.columns[i]} <-> {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.3f}")
else:
    print("Not enough numerical columns for correlation analysis")
```

## 6. Customer Demographics Analysis

```python
# Analyze demographic data
print("DEMOGRAPHIC DATASET ANALYSIS")
print("=" * 80)

# Get categorical columns in demographic data
demo_categorical = demographic_df.select_dtypes(include=['object']).columns
demo_numerical = demographic_df.select_dtypes(include=['int64', 'float64']).columns

print(f"Categorical columns: {list(demo_categorical)}")
print(f"Numerical columns: {list(demo_numerical)}")

# Show value distributions
for col in demo_categorical:
    print(f"\n{col.upper()} Distribution:")
    print("-" * 50)
    print(demographic_df[col].value_counts())
    print(f"Unique values: {demographic_df[col].nunique()}")
```

### 6.1 Demographic Visualizations

```python
# Visualize demographic distributions
if len(demo_categorical) > 0:
    n_cols = 2
    n_rows = (len(demo_categorical) + 1) // 2
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 5*n_rows))
    axes = axes.flatten() if len(demo_categorical) > 1 else [axes]
    
    for idx, col in enumerate(demo_categorical):
        if idx < len(axes):
            value_counts = demographic_df[col].value_counts()
            axes[idx].pie(value_counts.values, labels=value_counts.index, autopct='%1.1f%%', startangle=90)
            axes[idx].set_title(f'{col} Distribution', fontsize=12, fontweight='bold')
    
    # Hide empty subplots
    for idx in range(len(demo_categorical), len(axes)):
        axes[idx].axis('off')
    
    plt.tight_layout()
    plt.show()
```

## 7. Address Analysis

```python
# Analyze address data
print("ADDRESS DATASET ANALYSIS")
print("=" * 80)

addr_categorical = address_df.select_dtypes(include=['object']).columns
addr_numerical = address_df.select_dtypes(include=['int64', 'float64']).columns

for col in addr_categorical:
    unique_count = address_df[col].nunique()
    if unique_count < 50:  # Only show distributions for columns with reasonable number of unique values
        print(f"\n{col.upper()} - Top 10 Values:")
        print("-" * 50)
        print(address_df[col].value_counts().head(10))
    else:
        print(f"\n{col.upper()}: {unique_count} unique values (too many to display)")
        
print(f"\nNumerical columns in Address dataset: {list(addr_numerical)}")
```

## 8. Termination Analysis

```python
# Analyze termination data
print("TERMINATION DATASET ANALYSIS")
print("=" * 80)
print(f"Total terminations: {len(termination_df)}")
print(f"\nTermination dataset columns: {list(termination_df.columns)}")
print("\nFirst few records:")
display(termination_df.head(10))

term_categorical = termination_df.select_dtypes(include=['object']).columns
term_numerical = termination_df.select_dtypes(include=['int64', 'float64']).columns

# Analyze termination reasons or patterns
for col in term_categorical:
    print(f"\n{col.upper()} Distribution:")
    print("-" * 50)
    print(termination_df[col].value_counts())
    
# Statistical summary of termination data
print("\nStatistical Summary of Termination Data:")
display(termination_df.describe())
```

## 9. Key Insights and Summary

```python
# Summary of all datasets
print("=" * 80)
print("COMPREHENSIVE DATASET SUMMARY")
print("=" * 80)

summary_data = []
for name, df in datasets.items():
    summary_data.append({
        'Dataset': name,
        'Rows': len(df),
        'Columns': len(df.columns),
        'Missing Values': df.isnull().sum().sum(),
        'Duplicates': df.duplicated().sum(),
        'Memory Usage (MB)': df.memory_usage(deep=True).sum() / 1024**2
    })

summary_df = pd.DataFrame(summary_data)
display(summary_df)

print("\n" + "=" * 80)
print("DATA QUALITY INSIGHTS")
print("=" * 80)

for name, df in datasets.items():
    missing_pct = (df.isnull().sum().sum() / (df.shape[0] * df.shape[1])) * 100
    duplicate_pct = (df.duplicated().sum() / len(df)) * 100
    
    print(f"\n{name}:")
    print(f"  - Missing data: {missing_pct:.2f}%")
    print(f"  - Duplicate rows: {duplicate_pct:.2f}%")
    
    if missing_pct > 5:
        print(f"  ⚠️ High missing data - requires attention")
    if duplicate_pct > 1:
        print(f"  ⚠️ Contains duplicate records - may need deduplication")
```
